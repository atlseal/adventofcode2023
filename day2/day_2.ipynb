```python
# Part 1

games = {}
game_pattern = re.compile("Game (\d+):(.*)")
session_pattern = re.compile("([^;$]+)")
pulls_pattern = re.compile("(\d+) (\w+)")

max_of_color = {
    'red': 12,
    'green': 13,
    'blue': 14
}

file = open("input.txt", 'r')
# file = ["Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green",
# "Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue",
# "Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red",
# "Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red",
# "Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green"]
game_id_sum = 0

for line in file:
    game_match = game_pattern.match(line)
    game_id = int(game_match.group(1))
    game = game_match.group(2)
    
    sessions = session_pattern.findall(game)
    
    invalid_session = False
    for session in sessions:
        pulls = pulls_pattern.findall(session)
        for count, color in pulls:
            if max_of_color[color] < int(count):
                invalid_session = True
    if not invalid_session:
        game_id_sum += game_id

print(game_id_sum)
```

```python
# Part 2

games = {}
game_pattern = re.compile("Game (\d+):(.*)")
session_pattern = re.compile("([^;$]+)")
pulls_pattern = re.compile("(\d+) (\w+)")

file = open("input.txt", 'r')
# file = ["Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green",
# "Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue",
# "Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red",
# "Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red",
# "Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green"]
power_set_sum = 0

for line in file:
    game_match = game_pattern.match(line)
    game_id = int(game_match.group(1))
    game = game_match.group(2)
    
    sessions = session_pattern.findall(game)
    
    max_of_color = {
        'red': 0,
        'blue': 0,
        'green': 0
    }
    for session in sessions:
        pulls = pulls_pattern.findall(session)
        for count, color in pulls:
            max_of_color[color] = max(max_of_color[color], int(count))
    
    power_set_sum += max_of_color['red'] * max_of_color['blue'] * max_of_color['green']

print(power_set_sum)
```
