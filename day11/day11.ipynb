```python
# Part 1
input = open('input.txt','r')
lines = [line.strip() for line in input]

# lines = """...#......
# .......#..
# #.........
# ..........
# ......#...
# .#........
# .........#
# ..........
# .......#..
# #...#.....""".split('\n')

def expand_rows(lines):
    rows_to_dup = []
    
    for row_idx in range(len(lines)):
        if lines[row_idx].count('.') == len(lines[row_idx]):
            rows_to_dup.append(row_idx)
    
    while rows_to_dup:
        row_idx = rows_to_dup.pop()
        lines = lines[:row_idx] + ['.' * len(lines[row_idx])] + lines[row_idx:]
        
    return lines

def expand_cols(lines):
    cols_to_dup = []
    
    for col_idx in range(len(lines[0])):
        all_empty = True
        for row in lines:
            all_empty &= row[col_idx] == '.'
        if all_empty:
            cols_to_dup.append(col_idx)
            
    while cols_to_dup:
        col_idx = cols_to_dup.pop()
        for row_idx in range(len(lines)):
            lines[row_idx] = lines[row_idx][:col_idx] + '.' + lines[row_idx][col_idx:]
            
    return lines

def find(line, target):
    return [idx for idx, char in enumerate(line) if char == target]

def get_all_star_coords(lines):
    all_star_coords = []
    
    for row_idx, row in enumerate(lines):
        all_cols = find(row, '#')
        for col_idx in all_cols:
            all_star_coords.append((row_idx, col_idx))
            
    return all_star_coords

def dist(source, dest):
    return abs(source[0] - dest[0]) + abs(source[1] - dest[1])
            
def all_pairs(coords):
    all_coord_pairs = []
    
    for source_idx in range(len(coords)):
        for dest_idx in range(source_idx + 1, len(coords)):
            all_coord_pairs.append((coords[source_idx], coords[dest_idx], dist(coords[source_idx], coords[dest_idx])))
            
    return all_coord_pairs
    
lines = expand_rows(lines)
lines = expand_cols(lines)
star_coords = get_all_star_coords(lines)
all_pairs = all_pairs(star_coords)
print(sum([entry[2] for entry in all_pairs]))
```

```python
# Part 2
input = open('input.txt','r')
lines = [line.strip() for line in input]

# lines = """...#......
# .......#..
# #.........
# ..........
# ......#...
# .#........
# .........#
# ..........
# .......#..
# #...#.....""".split('\n')
# lines = """#.#""".split('\n')

def expand_rows(lines):
    rows_to_dup = []
    
    for row_idx in range(len(lines)):
        if lines[row_idx].count('.') == len(lines[row_idx]):
            rows_to_dup.append(row_idx)
    
    return rows_to_dup

def expand_cols(lines):
    cols_to_dup = []
    
    for col_idx in range(len(lines[0])):
        all_empty = True
        for row in lines:
            all_empty &= row[col_idx] == '.'
        if all_empty:
            cols_to_dup.append(col_idx)
            
    return cols_to_dup

def find(line, target):
    return [idx for idx, char in enumerate(line) if char == target]

def get_all_star_coords(lines):
    all_star_coords = []
    
    for row_idx, row in enumerate(lines):
        all_cols = find(row, '#')
        for col_idx in all_cols:
            all_star_coords.append((row_idx, col_idx))
            
    return all_star_coords

GROWTH_SIZE = 1000000

def dist(source, dest, rows_resized, cols_resized):
    global GROWTH_SIZE
    rows_traveled = set()
    cols_traveled = set()
#     print(f"rows_resized: {rows_resized}")
#     print(f"cols_resized: {cols_resized}")
    if source[0] >= dest[0]:
        rows_traveled.update([i for i in range(dest[0] + 1, source[0] + 1)])
    else:
        rows_traveled.update([i for i in range(source[0] + 1, dest[0] + 1)])
#     print(f"rows_traveled: {rows_traveled}")
        
    if source[1] >= dest[1]:
        cols_traveled.update([i for i in range(dest[1] + 1, source[1] + 1)])
    else:
        cols_traveled.update([i for i in range(source[1] + 1, dest[1] + 1)])
#     print(f"cols_traveled: {cols_traveled}")
        
    resized_rows_traveled = len(rows_traveled.intersection(rows_resized))
    standard_rows_traveled = len(rows_traveled) - resized_rows_traveled
    resized_cols_traveled = len(cols_traveled.intersection(cols_resized))
    standard_cols_traveled = len(cols_traveled) - resized_cols_traveled
#     print(resized_rows_traveled)
#     print(standard_rows_traveled)
#     print(resized_cols_traveled)
#     print(standard_cols_traveled)
    return GROWTH_SIZE * (resized_rows_traveled + resized_cols_traveled) + standard_rows_traveled + standard_cols_traveled
            
def all_pairs(rows_resized, cols_resized, coords):
    all_coord_pairs = []
    
    for source_idx in range(len(coords)):
        for dest_idx in range(source_idx + 1, len(coords)):
            all_coord_pairs.append((coords[source_idx], coords[dest_idx], dist(coords[source_idx], coords[dest_idx], rows_resized, cols_resized)))
            
    return all_coord_pairs
    
rows_idx_resized = expand_rows(lines)
cols_idx_resized = expand_cols(lines)
star_coords = get_all_star_coords(lines)
all_pairs = all_pairs(rows_idx_resized, cols_idx_resized, star_coords)
print(sum([entry[2] for entry in all_pairs]))
```
