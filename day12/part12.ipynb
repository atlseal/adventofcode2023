```python
# Part 1
input = open('input.txt','r')
lines = [line.strip() for line in input]

# lines = """???.### 1,1,3
# .??..??...?##. 1,1,3
# ?#?#?#?#?#?#?#? 1,3,1,6
# ????.#...#... 4,1,1
# ????.######..#####. 1,6,5
# ?###???????? 3,2,1""".split('\n')

def valid_string(line):
    combination, rules = line.split(' ')
    bad_groups = [len(x) for _, x in enumerate(combination.split('.')) if len(x) > 0]
    ordered_rules = [int(rule) for rule in rules.split(',')]
    return bad_groups == ordered_rules
    
def find_all(string, char):
    return [i for i, ltr in enumerate(string) if ltr == char]

def produce_replacement_string(length, pow_2_ident):
    return_val = ''
    while pow_2_ident > 0:
        next_char = '#' if pow_2_ident % 2 else '.'
        return_val = next_char + return_val
        pow_2_ident //= 2
    return '.' * (length - len(return_val)) + return_val

def get_combinations(line):
    combinations = 0
    all_qs = find_all(line, '?')
    total_combinations = 2**len(all_qs)
#     print(f"{line} Has {total_combinations} possible combinations")
    for i in range(total_combinations):
        new_str = ""
        replacement_str = produce_replacement_string(len(all_qs), i)
        str_components = line.split('?')
        for i in range(len(all_qs)):
            new_str += str_components[i] + replacement_str[i]
        new_str += str_components[-1]
        if valid_string(new_str):
            combinations += 1
#     print(f"{line} has {combinations} valid combinations")
    return combinations

# print(get_combinations(lines[1]))
sum = 0
lines_processed = 0
for line in lines:
    lines_processed += 1
    if lines_processed % 25 == 0:
        print(f"Processed {lines_processed} lines")
    sum += get_combinations(line)
print(sum)
```
