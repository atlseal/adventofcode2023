```python
# Part 1
import re

input = open("input.txt","r")
# input = """467..114..
# ...*......
# ..35..633.
# ......#...
# 617*......
# .....+.58.
# ..592.....
# ......755.
# ...$.*....
# .664.598..""".split('\n')

lines = [line.strip() for line in input]
# unique_characters = set()
# for line in lines:
#     for char in line:
#         unique_characters.add(char)
        
# print(unique_characters)

numbers_pattern = re.compile("\d+")
symbol_pattern = re.compile("[^\d.]")

def does_line_have_symbol(line_index, min_index, max_index):
    if line_index < 0 or line_index >= len(lines):
        return False
    
    min_index = max(min_index, 0)
    max_index = min(max_index, len(lines[line_index]))
    
    return symbol_pattern.search(lines[line_index][min_index:max_index])

sum = 0

for i in range(len(lines)):
    line = lines[i]
    numbers = numbers_pattern.finditer(line)
    for number in numbers:
        min_index = number.start()
        max_index = number.end()
        has_symbol_neighbor = does_line_have_symbol(i - 1, min_index - 1, max_index + 1) or \
                              does_line_have_symbol(i, min_index - 1, max_index + 1) or \
                              does_line_have_symbol(i + 1, min_index - 1, max_index + 1)
        if has_symbol_neighbor:
            sum += int(number.group(0))
            
print(sum)
```

```python
# Part 2
import re

input = open("input.txt","r")
# input = """467..114..
# ...*......
# ..35..633.
# ......#...
# 617*......
# .....+.58.
# ..592.....
# ......755.
# ...$.*....
# .664.598..""".split('\n')

lines = [line.strip() for line in input]

numbers_pattern = re.compile("\d+")
symbol_pattern = re.compile("\*")

def all_numbers_in_row(row):
    if row < 0 or row >= len(lines):
        return []
    
    return numbers_pattern.finditer(lines[row])

def find_neighboring_numbers(row, col):
    neighboring_numbers = []
    
    numbers_above = all_numbers_in_row(row - 1)
    numbers_on_same_row = all_numbers_in_row(row)
    numbers_below = all_numbers_in_row(row + 1)
    
    for number_match in numbers_above:
        start_idx = number_match.start()
        end_idx = number_match.end()
        value = int(number_match.group(0))
        
        if start_idx <= col and end_idx >= col or start_idx in (col, col + 1) or end_idx in (col, col + 1):
            neighboring_numbers += [value]
    
    for number_match in numbers_on_same_row:
        start_idx = number_match.start()
        end_idx = number_match.end()
        value = int(number_match.group(0))
        if start_idx == col + 1 or end_idx == col:
            neighboring_numbers += [value]
    
    for number_match in numbers_below:
        start_idx = number_match.start()
        end_idx = number_match.end()
        value = int(number_match.group(0))
        
        if start_idx <= col and end_idx >= col or start_idx in (col, col + 1) or end_idx in (col, col + 1):
            neighboring_numbers += [value]        
            
    return neighboring_numbers
    
sum = 0

for i in range(len(lines)):
    line = lines[i]
    gears = symbol_pattern.finditer(line)
    for gear in gears:
        gear_row = i
        gear_col = gear.start()
        neighboring_part_numbers = find_neighboring_numbers(gear_row, gear_col)
        if len(neighboring_part_numbers) == 2:
            sum += neighboring_part_numbers[0] * neighboring_part_numbers[1]
            
print(sum)
```
