```python
# Part 1
input = open('input.txt','r')

lines = [line.strip() for line in input.readlines()]
# lines = """.....
# .S-7.
# .|.|.
# .L-J.
# .....""".split('\n')
# lines = """7-F7-
# .FJ|7
# SJLL7
# |F--J
# LJ.LJ""".split('\n')
# lines = """L7..
# LS-7
# .L-J
# """.split('\n')

# Pad with '.' to be able to avoid boundary checking
lines = ['.' * len(lines[0])] + lines + ['.' * len(lines[0])]
lines = ['.' + line + '.' for line in lines]
# print("\n".join(lines))

# Find the s
s_row = None
s_col = None
for s_row in range(len(lines)):
    s_col = lines[s_row].find('S')
    if s_col != -1:
        break
        
# Since we are trying to find the furthest point, we follow all four possible directions until we get an overlapping step
steps = 1
pos_to_update = [(s_row - 1, s_col, 'N'), (s_row + 1, s_col, 'S'), (s_row, s_col - 1, 'W'), (s_row, s_col + 1, 'E')]
"""
| is a vertical pipe connecting north and south.
- is a horizontal pipe connecting east and west.
L is a 90-degree bend connecting north and east.
J is a 90-degree bend connecting north and west.
7 is a 90-degree bend connecting south and west.
F is a 90-degree bend connecting south and east.
"""
while True:
    found_loop = False
    for pos in pos_to_update:
        row, col,cur_direction = pos
        if lines[row][col] == '.':
            continue
        
        paired_pos_lookup = {
            '|': {'N':(row, col, 'S'), 'S':(row, col, 'N')},
            '-': {'E':(row, col, 'W'), 'W':(row, col, 'E')},
            'L': {'S':(row, col, 'W'), 'W':(row, col, 'S')},
            'J': {'S':(row, col, 'E'), 'E':(row, col, 'S')},
            '7': {'N':(row, col, 'E'), 'E':(row, col, 'N')},
            'F': {'N':(row, col, 'W'), 'W':(row, col, 'N')}
        }[lines[row][col]]
        if cur_direction not in paired_pos_lookup:
            continue
        found_loop |= paired_pos_lookup[cur_direction] in pos_to_update
    if found_loop: # Found a loop
        break
        
    steps += 1
    new_pos_to_check = []
    for pos in pos_to_update:
        row, col, cur_direction = pos
        space = lines[row][col]
        
        if space == '.': #Dead end
            continue
            
        new_pos_lookup = {
            '|': {'N':(row - 1, col, 'N'), 'S':(row + 1, col, 'S')},
            '-': {'E':(row, col + 1, 'E'), 'W':(row, col - 1, 'W')},
            'L': {'S':(row, col + 1, 'E'), 'W':(row - 1, col, 'N')},
            'J': {'S':(row, col - 1, 'W'), 'E':(row - 1, col, 'N')},
            '7': {'N':(row, col - 1, 'W'), 'E':(row + 1, col, 'S')},
            'F': {'N':(row, col + 1, 'E'), 'W':(row + 1, col, 'S')}
        }[lines[row][col]]
        if cur_direction not in new_pos_lookup:
            continue
        new_pos_to_check.append(new_pos_lookup[cur_direction])
    pos_to_update = new_pos_to_check
      
print(steps)
```

```python
# Part 2
from collections import defaultdict

input = open('input.txt','r')

lines = [line.strip() for line in input.readlines()]
# lines = """.....
# .S-7.
# .|.|.
# .L-J.
# .....""".split('\n')
# lines = """7-F7-
# .FJ|7
# SJLL7
# |F--J
# LJ.LJ""".split('\n')
# lines = """...........
# .S-------7.
# .|F-----7|.
# .||.....||.
# .||.....||.
# .|L-7.F-J|.
# .|..|.|..|.
# .L--J.L--J.
# ...........""".split('\n')
# lines = """..........
# .S------7.
# .|F----7|.
# .||....||.
# .||....||.
# .|L-7F-J|.
# .|..||..|.
# .L--JL--J.
# ..........""".split('\n')
# lines = """.F----7F7F7F7F-7....
# .|F--7||||||||FJ....
# .||.FJ||||||||L7....
# FJL7L7LJLJ||LJ.L-7..
# L--J.L7...LJS7F-7L7.
# ....F-J..F7FJ|L7L7L7
# ....L7.F7||L7|.L7L7|
# .....|FJLJ|FJ|F7|.LJ
# ....FJL-7.||.||||...
# ....L---J.LJ.LJLJ...""".split('\n')

# Pad with '.' to be able to avoid boundary checking
lines = ['.' * len(lines[0])] + lines + ['.' * len(lines[0])]
lines = ['.' + line + '.' for line in lines]

# Find the s
s_row = None
s_col = None
for s_row in range(len(lines)):
    s_col = lines[s_row].find('S')
    if s_col != -1:
        break
        
# Since we are trying to find the furthest point, we follow all four possible directions until we get an overlapping step
def find_path():
    paths_queue = [[(s_row - 1, s_col, 'N')], [(s_row + 1, s_col, 'S')], [(s_row, s_col - 1, 'W')], [(s_row, s_col + 1, 'E')]]
    """
    | is a vertical pipe connecting north and south.
    - is a horizontal pipe connecting east and west.
    L is a 90-degree bend connecting north and east.
    J is a 90-degree bend connecting north and west.
    7 is a 90-degree bend connecting south and west.
    F is a 90-degree bend connecting south and east.
    """
    while True:
        current_path = paths_queue.pop(0)
        row, col, direction = current_path[-1]
        if lines[row][col] == '.':
            continue
        
        if lines[row][col] == 'S':
            return set(map(lambda x: (x[0], x[1]), current_path))
        
        new_pos_lookup = {
                '|': {'N':(row - 1, col, 'N'), 'S':(row + 1, col, 'S')},
                '-': {'E':(row, col + 1, 'E'), 'W':(row, col - 1, 'W')},
                'L': {'S':(row, col + 1, 'E'), 'W':(row - 1, col, 'N')},
                'J': {'S':(row, col - 1, 'W'), 'E':(row - 1, col, 'N')},
                '7': {'N':(row, col - 1, 'W'), 'E':(row + 1, col, 'S')},
                'F': {'N':(row, col + 1, 'E'), 'W':(row + 1, col, 'S')}
            }[lines[row][col]]
        if direction not in new_pos_lookup:
            continue
            
        next_pos = new_pos_lookup[direction]
        current_path.append(next_pos)
        paths_queue.append(current_path)
        
def clear_non_pipe_spots(pipe_path):
    for row_idx in range(len(lines)):
        for col_idx in range(len(lines[0])):
            if (row_idx, col_idx) not in pipe_path:
                lines[row_idx] = lines[row_idx][:col_idx] + '.' + lines[row_idx][col_idx + 1:]
    
def fill_interior():
    for row_idx in range(0, len(lines)):
        for col_idx in range(0, len(lines[0])):
            if lines[row_idx][col_idx] != '.':
                continue
                
            # Draw line in all 4 directions, if odd crossings in all four, is interior
            # Need to consider straight line of points as a single crossing for this to work
            if is_interior(row_idx, col_idx):
                lines[row_idx] = lines[row_idx][:col_idx] + 'I' + lines[row_idx][col_idx + 1:]
                
def is_interior(row, col):
    above = count_pipes_above(row, col)
    if above % 2 != 1:
        return False
    
    below = count_pipes_below(row, col)
    if below % 2 != 1:
        return False
    
    left = count_pipes_left(row, col)
    if left % 2 != 1:
        return False
    
    right = count_pipes_right(row, col)
    if right % 2 != 1:
        return False
    
    return True
            
def count_pipes_above(row, col):
    pipes = 0
    last_corner = None
    
    for offset in range(1, row+1):
        if lines[row - offset][col] in ('.','|','I'):
            continue
        elif lines[row - offset][col] == '-':
            pipes += 1
        elif last_corner:
            # Since we're only going up, we are ensured that a lower corner
            # is the last corner
            if last_corner == 'L' and lines[row - offset][col] == '7':
                pipes += 1
            elif last_corner == 'J' and lines[row - offset][col] == 'F':
                pipes += 1
            last_corner = None
        else:
            last_corner = lines[row - offset][col]

    return pipes

def count_pipes_below(row, col):
    pipes = 0
    last_corner = None
    
    for offset in range(1, len(lines) - row):
        if lines[row + offset][col] in ('.','|','I'):
            continue
        elif lines[row + offset][col] == '-':
            pipes += 1
        elif last_corner:
            # Since we're only going down, we are ensured that an upper corner
            # is the last corner
            if last_corner == '7' and lines[row + offset][col] == 'L':
                pipes += 1
            elif last_corner == 'F' and lines[row + offset][col] == 'J':
                pipes += 1
            last_corner = None
        else:
            last_corner = lines[row + offset][col]

    return pipes

def count_pipes_left(row, col):
    pipes = 0
    last_corner = None
    
    for offset in range(1, col+1):
        if lines[row][col - offset] in ('.','-','I'):
            continue
        elif lines[row][col - offset] == '|':
            pipes += 1
        elif last_corner:
            # Since we're only going left, we are ensured that a right corner
            # is the last corner
            if last_corner == '7' and lines[row][col - offset] == 'L':
                pipes += 1
            elif last_corner == 'J' and lines[row][col - offset] == 'F':
                pipes += 1
            last_corner = None
        else:
            last_corner = lines[row][col - offset]

    return pipes

def count_pipes_right(row, col):
    pipes = 0
    last_corner = None
    
    for offset in range(1, len(lines[0]) - col):
        if lines[row][col + offset] in ('.','-','I'):
            continue
        elif lines[row][col + offset] == '|':
            pipes += 1
        elif last_corner:
            # Since we're only going right, we are ensured that a left corner
            # is the last corner
            if last_corner == 'L' and lines[row][col + offset] == '7':
                pipes += 1
            elif last_corner == 'F' and lines[row][col + offset] == 'J':
                pipes += 1
            last_corner = None
        else:
            last_corner = lines[row][col + offset]
    
    return pipes

def replace_s(pipe_path):
    for pipe in pipe_path:
        row, col = pipe
        if lines[row][col] != 'S':
            continue
            
        connected_above = (row - 1, col) in pipe_path and lines[row - 1][col] in ('|','7','F')
        connected_below = (row + 1, col) in pipe_path and lines[row + 1][col] in ('|','L','J')
        connected_left = (row, col - 1) in pipe_path and lines[row][col - 1] in ('-','L','F')
        connected_right = (row, col + 1) in pipe_path and lines[row][col + 1] in ('-', 'J', '7')
            
        symbol = None
        if connected_above and connected_below:
            symbol = '|'
        elif connected_left and connected_right:
            symbol = '-'
        elif connected_above and connected_right:
            symbol = 'L'
        elif connected_left and connected_above:
            symbol = 'J'
        elif connected_left and connected_below:
            symbol = '7'
        elif connected_below and connected_right:
            symbol = 'F'
        else:
            print("SHOULDN'T GET HERE")
            
        lines[row] = lines[row][:col] + symbol + lines[row][col + 1:]
        break

##################

pipe_path = find_path()
clear_non_pipe_spots(pipe_path)
replace_s(pipe_path)
fill_interior()
# print('\n'.join(lines))
print(sum([line.count('I') for line in lines]))
```
